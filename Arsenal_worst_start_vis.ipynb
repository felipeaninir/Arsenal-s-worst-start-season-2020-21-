```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename = '/content/drive/MyDrive/PTDataScience/Visualization/Week51.xlsx'
df = pd.read_excel(filename)
print(df.shape)
df
```

```python
plt.rcParams["figure.figsize"] = (20, 20)  
fig, axes = plt.subplots(2, 2) 
fig.suptitle('Arsenal results after 10 games in the Premier League (since 1992/93)', fontsize=20)

axes[0,0].set_title('Table position', fontsize=20)
axes[0,1].set_title('Points', fontsize=20)
axes[1,0].set_title('Goals scored', fontsize=20)
axes[1,1].set_title('Goals diffrence', fontsize=20)

axes[0,0].barh(df['Season'].values, df['Position'].values)
axes[0,1].barh(df['Season'].values, df['Points'].values)
axes[1,0].barh(df['Season'].values, df['Goals For'].values)
axes[1,1].barh(df['Season'].values, df['Goal Difference'].values)

axes[0,0].barh(df['Season'][28], df['Position'][28], color='r')
axes[0,1].barh(df['Season'][28], df['Points'][28], color='r')
axes[1,0].barh(df['Season'][28], df['Goals For'][28], color='r')
axes[1,1].barh(df['Season'][28], df['Goal Difference'][28], color='r')

axes
```
